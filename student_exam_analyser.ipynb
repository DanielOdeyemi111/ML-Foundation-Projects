# Student Exam Analyser
# Analyzing 30 students exam and grads using Pandas and Numpy
# Libraries Pandas, Numpy

```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
# Generating Data
np.random.seed(42)
random_scores = np.random.randint(40, 100, size = 30)
random_scores
student_name = np.array("Ade")
student_name
```

```python
# Generating DataFrame from Data
student_data = pd.DataFrame({"Names" : student_name, "Scores" : random_scores})
student_data
```

```python
# Generating Data Mean,Min, Max and Standard Deviation
student_data["Scores"].mean()
student_data["Scores"].min()
student_data["Scores"].max()
student_data["Scores"].std()
```

```python
# Generating assign grades function
def assign_grades (score):
    if score >= 80:
        return "A"
    elif score >= 65 and score < 80:
        return "B"
    elif score >= 50 and score < 65:
        return "C"
    elif score < 50:
        return "F"
    
```

```python
#Generating grades on the Dataframe from the assign grades function
student_data["Grades"] = student_data["Scores"].apply(assign_grades)
student_data
```

```python
# Generating the grade counter
grade_count = student_data["Grades"].value_counts().sort_index()
grade_count
```
